```python
def rgb_to_hex(r, g, b):
    r, g, b = int(r), int(g), int(b)
    return '#' + hex(r)[2:].zfill(2) + hex(g)[2:].zfill(2) + hex(b)[2:].zfill(2)


f_c = open("elements.ini",'r')
line = f_c.readlines()
#print(line)
f_c.close()


elements_split = []

for i in line :
    elements_split.append(i.split())
    
    
color_code = []    
for i in elements_split:
    color_code.append([i[1],rgb_to_hex(int(255*float(i[5])),int(255*float(i[6])),int(255*float(i[7])))])
    
print(color_code)
```

# XRD pattern

```python
def XRD(name_file,start,end,color,name="") :
    
    f_XRD = open(name_file,'r')
    lines = f_XRD.readlines()
    XRD = list()
    for i,line in enumerate(lines):
        line = line.split()  # 줄 끝의 줄 바꿈 문자를 제거한다.
        XRD.append(line)
    f_XRD.close()
    
    XRD = XRD[2+(start-1)*100:3+(end-1)*100]
    #print(XRD)
    
    df_XRD = pd.DataFrame(XRD)
    df_XRD.columns = ['2θ','Intensity','remarks']
    
    fig,ax  = plt.subplots(figsize=(8,6))


    degree=df_XRD['2θ']
    degree= list(map(float,degree))
    intensity = df_XRD['Intensity']
    intensity= list(map(float,intensity))

    
    ax.plot(degree,intensity,c=color,linewidth=2)


    ax.set_yticks([])
    ax.set_xticks(np.arange(start,end+0.01,10))
    plt.xticks(fontproperties=font_ticks)
    ax.set_xlabel("2θ (degree)", fontproperties=font_labels)
    ax.set_ylabel("Intensity", fontproperties=font_labels)

    ax.set_xlim([start,end])
    
    ax.text(start+(start+end)/50,95,name,fontproperties=font_text)
    
    return df_XRD
```

```python
c= ['#02DAA2','#00b3fe','#0E8BB3','#120458']

def XRD_series(name_file,start,end,color,name="") :
    fig,axes  = plt.subplots(len(name_file),1,sharex=True,figsize=(8,8)) # x를 공유하는 len(name_file) 행, 1열의 subplots 들 생성  
    for i in range(len(name_file)):
    
        f_XRD = open(name_file[i],'r')
        lines = f_XRD.readlines()
        XRD = list()
        for j,line in enumerate(lines):
            line = line.split()  
            XRD.append(line)
        f_XRD.close()
    
        XRD = XRD[2+(start-1)*100:3+(end-1)*100]
    #print(XRD)
    
        df_XRD = pd.DataFrame(XRD)
        df_XRD.columns = ['2θ','Intensity','remarks']

        degree=df_XRD['2θ']
        degree= list(map(float,degree))
        intensity = df_XRD['Intensity']
        intensity= list(map(float,intensity))
        
        axes[i].plot(degree,intensity,c=color[i],linewidth=2)

        axes[i].set_ylim([-10,100])

        axes[i].set_yticks([])
        axes[i].set_xticks(np.arange(start,end+0.01,10))
        plt.xticks(fontproperties=font_ticks)
        
        axes[i].set_xlim([start,end])
        axes[i].text(start+(start+end)/50,75,name[i],fontproperties=font_text)
        
        plt.subplots_adjust( hspace=0) #subplot 간 간격조절
        
        # subplot 사이의 축 제거
        if i != 0 :
            axes[i].spines['top'].set_visible(False)
        
        if i != len(name_file)-1 :
            axes[i].spines['bottom'].set_visible(False)
        
    # 전체에 대한 x,y label 설정
    fig.text(0.5, 0.06, '2θ (degree)', ha='center', fontproperties=font_labels)
    fig.text(0.06, 0.5, 'Intensity', va='center', rotation='vertical', fontproperties=font_labels)
    plt.show()
        
```

```python
stages = ["stage1.int","stage2.int","stage3.int","stage4.int"]

XRD_series(stages,20,40,c,['stage1','stage2','stage3','stage4']) 
```

```python
from scipy import stats

#from scipy.optimize import curve_fit

#def func(x, a, b):
#    return a*x + b

plt.figure(figsize=(8,6))

G_OH, G_O = df['G_OH'],df['G_O']

plt.scatter(G_OH, G_O ,label = 'data' ,color=c[0], alpha=0.5 ,s=100,marker='s')

# 1차식 형태로 fitting
res = stats.linregress(G_OH, G_O)
x_fit = np.arange(-1,6,0.01)
y_fit = res.intercept + res.slope* x_fit
plt.plot(x_fit,y_fit,'--',label='fitted line',c='black',linewidth=1.5,alpha=0.5)


plt.xlim(-.5,3)
plt.ylim(-1,8)

plt.xticks(fontproperties=font_ticks )
plt.yticks(fontproperties=font_ticks )


plt.xlabel('$ \mathrm{ΔG_{OH}(eV)} $',fontproperties=font_labels) 
plt.ylabel('$ \mathrm{ΔG_{O}(eV)} $',fontproperties=font_labels)


plt.text( -0.4,7 ,'$ \mathrm{ΔG_{O}  = %.2f ΔG_{OH}  + %.2f} $' %(res.slope, res.intercept),fontproperties=font_text)
plt.text( -0.4,6.5 ,'$ \mathrm{R^2 = %.2f}$' %res.rvalue**2,fontproperties=font_text)

plt.legend(fontsize=14)
```

```python
def reaction_pathway_with_barrier(step_list,ylim,c,lab,barrier=False,yaxis='Relative energy (eV)') :
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    
    for i in range(len(step_list)) :
        
        rc = np.arange(len(step_list[i]) * 2)
        data = {'rc' : rc}
        df = pd.DataFrame(data)
        dash = list()
        #print(step_list)
        for j in range(len(step_list[i])) :
            df['s'+str(j)] = np.nan
            df['s'+str(j)][2*j] = step_list[i][j]
            df['s'+str(j)][2*j+1] = step_list[i][j]
            dash.append(step_list[i][j])
            dash.append(step_list[i][j])

        df['dash'] = dash    
        
        ax.plot([0,1],[0,0],color = c[i],label = lab[i])

        ax.plot(df.loc[:, 's0':'s'+str(len(step_list[i])-1)],color = c[i])
        
        
        # 1 3 5 7 --> 1 2 3 4
        for k in range(1,len(df.index)-1,2) :
            check_dash = True
            if barrier != False and len(barrier[i]) != 0 :
                for l in range(len(barrier[i][0])) :                  
                    if barrier[i][0][l] == k//2 + 1 :
                        
                        x1 = np.arange(0,np.pi/2,0.01)
                        y1 = barrier[i][1][l] * np.sin(x1)

                        x1_set = x1/(np.pi/2) * 0.5 + k
                        y1_set = y1 + dash[k]

                        x2 = np.arange(np.pi/2,np.pi,0.01)
                        y2 = (barrier[i][1][l]-(dash[k+1]-dash[k])) * np.sin(x2)

                        x2_set = (x2-(np.pi/2))/(np.pi/2) * 0.5 + k+0.5
                        y2_set = y2 +dash[k+1]

                        ax.plot(x1_set,y1_set,c[i])
                        ax.plot(x2_set,y2_set,c[i])

                        check_dash = False
                        
                        

            if check_dash == True :
                ax.plot([k,k+1],[dash[k],dash[k+1]],c[i],linestyle = 'dashed')
                        
                
            if barrier == False :
                for k in range(1,len(df.index)-1,2) :
                    ax.plot([k,k+1],[dash[k],dash[k+1]],c[i],linestyle = 'dashed')
    
    #plt.plot(df.index,df[ 'dash' ],c,linestyle = 'dashed')
    
    plt.xticks([])
    plt.yticks(fontproperties=font_ticks)
    ax.set_ylim(ylim)
    
    ax.set_xlabel('Reaction coordinates', fontproperties=font_labels)
    ax.set_ylabel(yaxis, fontproperties=font_labels)
    font_name = fm.FontProperties(fname=font_path).get_name()
    plt.legend(prop={'family':font_name, 'size':14},frameon=False)
    
```

```python
rxn1 = [0,1.23,1.23*2,1.23*3,1.23*4]
rxn2 = [0,1.,3,3.5,1.23*4]

rxn = [rxn1,rxn2]
y_lim = [-1,6]
label = ['rxn1','rxn2']

reaction_pathway_with_barrier(rxn,y_lim,c,label)
```

```python
barrier1 = [[2,3],[3,3]]  # 2,3번째 step 에 각각 3,3 의 barrier
barrier2 = [[2,4],[4,2]]  # 2,4번째 step 에 각각 4,2 의 barrier

barrier = [barrier1,barrier2]
reaction_pathway_with_barrier(rxn,y_lim,c,label,barrier,'$ \mathrm{ΔG (eV)}$')
```
